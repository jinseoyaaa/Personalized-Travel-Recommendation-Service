```python
import platform 
from matplotlib import font_manager, rc
from collections import Counter
import matplotlib.pyplot as plt
import squarify

if platform.system() == 'Windows': 
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else: 
    print('Check your OS system')
```

```python
words = '태조산 원사 독립기념관 목천읍 천안역 성남면 안지원 신부동 유관순 병천 아우 유량 겨레 사건번호 벚꽃 먹다 신방동 열사 물건 성천 북일고 독립 대호수 토지 흑성산 호두과자 체험 병천면 카페 명소 광덕면 좋다 원리 북면 사담 맛있다 안전 예쁘다 대지 월건 흑성산성 노벨 청동'
word_list = words.split()
```

```python
word_counts = Counter(word_list)
```

```python
# 단어와 빈도수 리스트
words = list(word_counts.keys())
counts = list(word_counts.values())

# 트리맵 시각화
plt.figure(figsize=(8, 6))
squarify.plot(sizes=counts, label=words, alpha=0.8, linewidth=1, edgecolor='white')
plt.title('충청남도 천안시')
plt.axis('off')
plt.show()
plt.savefig('image/44130.jpg', dpi=500)
```
