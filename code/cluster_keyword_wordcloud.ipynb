```python
import platform
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
if platform.system() == 'Windows':  # 윈도우의 경우
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == "Darwin":  # Mac의 경우
    font_path = "/Users/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
df = pd.read_csv('data/전국_시군구_키워드_군집.csv', encoding='cp949')
```

```python
cluster_0 = df[df['cluster'] == 0]
```

```python
cluster_1 = df[df['cluster'] == 1]
```

```python
cluster_2 = df[df['cluster'] == 2]
```

```python
cluster_3 = df[df['cluster'] == 3]
```

```python
cluster_4 = df[df['cluster'] == 4]
```

```python
keywords_total = []

for keywords in cluster_4['keyword']:  
    keywords_list = keywords.split(' ')
    for keyword in keywords_list:
        keywords_total.append(keyword)
```

```python
keywords_counts = Counter(keywords_total)
```

```python
keywords_counts.most_common(200)
```

```python
wordcloud = WordCloud(font_path= font_path,  # 사용할 글꼴 경로 
                    background_color="white",  # 배경색
                    max_words=100,  # 최대 몇 개의 단어를 나타낼 것인지 설정
                    relative_scaling= 0.3,  # 워드클라우드 내 글자들의 상대적인 크기(0~1)
                    width = 800,  # 워드클라우드 가로
                    height = 400  # 워드클라우드 세로
                 ).generate_from_frequencies(keywords_counts)

plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('image/cluster_4.jpg')  # 파일명 설정
```
